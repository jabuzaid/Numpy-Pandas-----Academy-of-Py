```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# District Summary as follows:

# Calculate the total number of schools
school_count = school_data["school_name"].count()

# Calculate the total number of students
student_count = school_data_complete["student_name"].count()

# Calculate the total budget
total_budget = school_data["budget"].sum()

# Calculate the average math score
avg_math_score = school_data_complete["math_score"].mean()

# Calculate the average reading score
avg_reading_score= school_data_complete["reading_score"].mean()

# Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
overall_passing_rate = (avg_math_score + avg_reading_score)/2
overall_pass_percent = overall_passing_rate / student_count * 100 

# Calculate the percentage of students with a passing math score (70 or greater)
filtered_math_df = school_data_complete[school_data_complete["math_score"] >= 70]
math_pass_percent = filtered_math_df["math_score"].count() / student_count * 100

# Calculate the percentage of students with a passing reading score (70 or greater)
filtered_reading_df = school_data_complete[school_data_complete["reading_score"] >= 70]
reading_pass_percent = filtered_reading_df["reading_score"].count() / student_count * 100
```

```python
# Create a dataframe to hold the above results in a cleaner format.
dis_sum_df = pd.DataFrame({"Total Schools": [school_count], 
           "Total Students": [student_count], 
           "Total Budget": [total_budget], 
           "Average Math Score": [avg_math_score],
            "Average Reading Score": [avg_reading_score],
            "% Passing Math":[math_pass_percent],
            "% Passing Reading":[reading_pass_percent],
            "% Overall Passing Rate":[overall_passing_rate ]})

# Formatting the cell results as required
dis_sum_df["Total Students"] = dis_sum_df["Total Students"].map("{:,}".format)
dis_sum_df["Total Budget"] = dis_sum_df["Total Budget"].map("${:,.2f}".format)
dis_sum_df
```

```python
# Writing the district result to an output file
dis_sum_df.to_csv("Output/PyCityDistrictAnalyses.csv", index=False, header=True)
```

```python
# Creating an overview table that summarizes key metrics FOR each school, including:

# School Name
grouped_school_df = school_data_complete.groupby(["school_name"])

# School Type
SchoolType = grouped_school_df["type"].first()

# Total Students
TotalStudents = grouped_school_df["Student ID"].count()

# Total School Budget
TotalSchoolBudget = grouped_school_df["budget"].first()

# Per Student Budget
PerStudentBudget = TotalSchoolBudget / TotalStudents

# Average Math Score
avg_math_score = grouped_school_df["math_score"].mean()

# Average Reading Score
avg_reading_score = grouped_school_df["reading_score"].mean()

# % Passing Math
pass_math_score = school_data_complete[school_data_complete["math_score"] >= 70].groupby(["school_name"])["math_score"].count()
pass_math_percent = pass_math_score/TotalStudents * 100

# % Passing Reading
pass_reading_score = school_data_complete[school_data_complete["reading_score"] >= 70].groupby(["school_name"])["reading_score"].count()
pass_reading_percent = pass_reading_score/TotalStudents * 100

#Overall Passing Rate (Average of the above two)
overall_pass_rate = (pass_math_percent + pass_reading_percent) / 2

# Create a dataframe to hold the above results
school_data_df = pd.DataFrame({"School Type": SchoolType,
      "Total Students": TotalStudents,
      "Total School Budget": TotalSchoolBudget,
      "Per Student Budget": PerStudentBudget,
      "Average Math Score": avg_math_score,
      "Average Reading Score": avg_reading_score,
      "% Passing Math": pass_math_percent,
      "% Passing Reading": pass_reading_percent,
      "% Overall Passing Rate": overall_pass_rate})

# Formatting the cell results as required
school_data_df["Total School Budget"] = school_data_df["Total School Budget"].map("${:,.2f}".format)
school_data_df["Per Student Budget"] = school_data_df["Per Student Budget"].map("${:,.2f}".format)
school_data_df
```

```python
# Writing the school analyses to an output file
school_data_df.to_csv("Output/PyCitySchooltAnalyses.csv", index=False, header=True)
```
